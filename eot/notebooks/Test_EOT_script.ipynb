```python
# read data from text files
with open('data/eot_sentences_test.txt', 'r') as f:
    utterances_test = f.read()
with open('data/eot_target_test.txt', 'r') as f:
    labels_test = f.read()
```

```python
from string import punctuation

def split_questions (question):
    
    # get rid of punctuation
    question = question.lower() # lowercase, standardize
    all_text = ''.join([c for c in question if c not in punctuation])

    # split by new lines and spaces
    reviews_split = all_text.split('\n')
    all_text = ' '.join(reviews_split)

    # create a list of words
    words = all_text.split()
    
    return reviews_split
```

```python
utterances_split = split_questions(utterances_test)
```

```python
def encode_labels (labels):
    labels_split = labels.split('\n')
    encoded_labels = np.array([1 if label == 'complete' else 0 for label in labels_split])
    return encoded_labels

labels_test = encode_labels(labels_test)
```

```python
def remove_outliers(reviews_ints, encoded_labels):

    print('Number of reviews before removing outliers: ', len(reviews_ints))
    
    ## remove any reviews/labels with zero length from the reviews_ints list.

    # get indices of any reviews with length 0
    non_zero_idx = [ii for ii, review in enumerate(reviews_ints) if len(review.strip()) != 0]

    # remove 0-length reviews and their labels
    reviews_ints = [reviews_ints[ii] for ii in non_zero_idx]
    encoded_labels = np.array([encoded_labels[ii] for ii in non_zero_idx])
    
    print('Number of reviews after removing outliers: ', len(reviews_ints))
    
    return reviews_ints, encoded_labels
```

```python
utterances_final, labels_final = remove_outliers(utterances_split, labels_test)

print(len(utterances_final), len(labels_final))

pd.DataFrame(data = {'utterance': utterances_final, 'tatget': labels_final})
```

### Testing

```python
def test_accuracy():
    complete = 0
    incomplete = 0
    result = 0
    
    for i, utterance in enumerate(utterances_final):

        utt_list = utterance.split()
        res = len(utt_list) 
        last_word = utt_list[-1]

        prediction = check(last_word, antecedents)

        if prediction == labels_final[i]:
            complete = complete + 1
        else:
            incomplete = incomplete + 1
            


        if prediction != labels_final[i]:
            print(prediction, labels_final[i])
            print(utterance)
            
    result = complete / (complete+incomplete)
    assert(len(labels_final) == complete+incomplete)
    print("The accuracy is: {}".format(result))
    
   
```

```python
test_accuracy()
```
